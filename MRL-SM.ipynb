# MODELO DE REGRESIÓN LINEAL

```python
import warnings
warnings.filterwarnings('ignore')

#Librerias del Modelo 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
import sklearn.metrics as metrics
! wget https://storage.googleapis.com/aerial-yeti-350323.appspot.com/DataMRL.csv
```

```python
#Cargamos los resultados de estudios de pacientes de nuestra base de datos
datos = pd.read_csv('DataMRL.csv')
```

```python
#Mostramos el dataset y sus columnas
print(datos.columns)
print('\n')
datos.head(5)
```

# SINDROME METABOLICO - ATTP III 

```python
#Identificamos las variables DEPENDIENTES 
X=datos.iloc[:, 1:12]
#Mostramos como variable INDEPENDIENTES
y = datos.iloc[:,25]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.8)
logit_model = linear_model.LogisticRegression()
logit_model.fit(X_train,y_train)
print("el ajuste es: ",logit_model.score(X_train,y_train))
print("el intercepto (b0): ",logit_model.intercept_)
print("los coeficientes(b0): ",logit_model.coef_)
```

```python
#Entrada de los datos para tener una prediccion del Sindrome Metabolico por la ATTP III
#	['Sexo', 'Edad', 'Peso', 'Glucosa ', 'Trigliceridos', 'Colesterol', 'HDL', 'LDL', 'IMC', 'ICC', 'Talla']
pred = [0, 39, 78.9, 140, 206, 206, 37, 147, 34.15, 0.86, 1.52]
```

```python
prediccion = logit_model.coef_ * pred
```

```python
respuesta = prediccion.sum() + logit_model.intercept_
```

```python
if respuesta <= 0:
  print('Baja Probabilidad')
if respuesta >= 0:
  print('Alta Probabilidad')
```

# Dislipidemias

```python
def Dislipidemias():

  Colesterol = int(input('Ingresa tus datos de Colesterol: '))
  Trigliceridos = int(input('Ingresa tus datos de Trigliceridos: '))
  LDL = int(input('Ingresa tus datos de LDL: '))
  HDL = int(input('Ingresa tus datos de HDL: '))

  
  
  if Colesterol > 200 and Trigliceridos > 200 and LDL >= 130:
    print('\n', 'Hipercolesterolemia')
  elif Colesterol < 200 and Trigliceridos > 150 and LDL < 130:
    print('\n', 'Hipertrigliceridemia')
  elif Colesterol > 200 and Trigliceridos > 150 and LDL >= 130 or HDL >= 160:
    print('\n', 'Hiperlipidemia Mixta')
  elif HDL < 40:
    print('\n', 'Hipoalfalipoproteinemia')  
```
